```python
import selenium
import pandas as pd
from selenium import webdriver
```

```python
driver1 = webdriver.Chrome("chromedriver.exe")
```

```python
url = 'https://www.cardekho.com/best-sedan'
driver1.get(url)
```

```python
carname = driver1.find_elements_by_xpath("//div[@class='gsc_col-xs-12 holder truncate']")
carname
```

```python
car_names=[]
for i in carname:
    car_names.append(i.text)
car_names
```

```python
carprice = driver1.find_elements_by_xpath("//div[@class='price']")
carprice
```

```python
car_price=[]
for i in carprice:
    car_price.append(i.text)
car_price
```

```python
import pandas as pd
cars=pd.DataFrame({})
cars['name']=car_names
cars['price']=car_price
cars
```
